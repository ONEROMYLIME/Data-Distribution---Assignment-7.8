```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

##  Gumbel distribution
gumbel = pd.DataFrame()
gumbel = np.random.gumbel(0, 1, 1000)

plt.hist(gumbel)
plt.axvline(gumbel.mean(), color='pink', linestyle='solid', linewidth=2)
plt.axvline(gumbel.mean() + gumbel.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
# This does an accurate way of depicting the standard devation and the mean as well as the range between the two to depict the probability of an event but again depending on the data may vary.
```

```python
## Gamma Distribution
gamma = pd.DataFrame()
gamma = np.random.gamma(0, 1, 1000)

plt.hist(gumbel)
plt.axvline(gamma.mean(), color='pink', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
## In this distribution both the mean and standard deviation overlap at one of the highest points of an event.
```

```python
## Multinomial Distribution

multinomial = pd.DataFrame()
multinomial = np.random.multinomial(1, [0, 1000] )

plt.hist(multinomial)
plt.axvline(multinomial.mean(), color='pink', linestyle='solid', linewidth=2)
plt.axvline(multinomial.mean() + multinomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(multinomial.mean() - multinomial.std(), color='b', linestyle='dashed', linewidth=2)


plt.show()
```

```python
## in the multinomial distribution the outcomes are very distant and so the mean is right in the middle but the standard deviation is on the highest and lowest points
```

```python
## Negative Binomial Distribution
negative_binomial = pd.DataFrame()
negative_binomial = np.random.negative_binomial(1, 0, 1000)

plt.hist(negative_binomial)
plt.axvline(negative_binomial.mean(), color='pink', linestyle='solid', linewidth=2)
plt.axvline(negative_binomial.mean() + negative_binomial.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
## the negative binomial overlaps but has no range
```

```python
## Poisson Distribution

poisson = pd.DataFrame()
poisson = np.random.poisson(1, 1000)

plt.hist(poisson)
plt.axvline(poisson.mean(), color='pink', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() + poisson.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean() - poisson.std(), color='b', linestyle='dashed', linewidth=2)

plt.show()
```

```python
#  varies by data sometimes the mean is at the highest point of the range
```

```python
## Additonally

star1 = np.random.normal(5,.5, 1000)
star2 = np.random.normal(10,1, 1000)
star3 = star1+star2

plt.hist(star3)
plt.axvline(star3.mean(), color='pink', linestyle='solid', linewidth=2)
plt.axvline(star3.mean() + star3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(star3.mean() - star3.std(), color='b', linestyle='dashed', linewidth=2)


plt.show()

print("slight skew before and after the mean depending on the data ")
```
